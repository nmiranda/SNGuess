```python
import requests
import pandas as pd
import io
```

## Steps to query for RCF/BTS candidate list

```python
query = "https://sites.astro.caltech.edu/ztf/bts/explorer.php?f=s&format=csv"
```

```python
req = requests.get(query)
bts_transients = pd.read_csv(io.StringIO(req.text), index_col='ZTFID')
```

```python
bts_transients.to_csv('bts_transients.csv')
```
